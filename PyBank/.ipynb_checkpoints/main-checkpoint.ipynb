```python
from pathlib import Path
import csv
#print(Path.cwd())
# @TODO: Set file paths for budget_data.csv

budget_data = Path('C:/Users/MOMO/desktop/python_homework/PyBank/budget_data.csv')
budget_data_out = Path('C:/Users/MOMO/desktop/python_homework/PyBank/budget_analysis.txt')

#et varaiables to hold lists 
list_of_pnl= [] 
net_change_list = [] 
month_of_change = [] 
greatest_increase = 0
greatest_decrease = 0
total_months = 0
total_pnl = 0
total_net = 0 
line_num = 0 

#open file with csv_reader
with open(budget_data) as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',')
     #read header   
    csv_header = next(csv_reader)
     #print(row)
 # append  profit/loss, and months as lists
    for row in csv_reader:
        line_num += 1
        total_net += int(row[1])
        list_of_pnl.append(int(row[1]))
        month_of_change.append(row[0])
        

net_change_list = []       
for i in range(1, len(list_of_pnl)):
    netchange = list_of_pnl[i] - list_of_pnl[i - 1]
    net_change_list.append(int(netchange))   


for i in range(0, len(net_change_list)):                  
    total_pnl = sum(net_change_list) 
    average_pnl = round((total_pnl / (len(net_change_list))), 2)

    
for pnl in list_of_pnl:
   # total_pnl =+ list_of_pnl(int(row[1])
    if greatest_decrease == 0:
        greatest_decrease= pnl
        greatest_increase= pnl
    elif pnl < greatest_decrease:
        greatest_decrease = pnl
    elif pnl > greatest_increase:
        greatest_increase = pnl
#identify the index of max and min changes

#find the index holding highest and lowest changes and call the months
greatest_increase_pl = list_of_pnl.index(greatest_increase)
greatest_decrease_pl = list_of_pnl.index(greatest_decrease)

greatest_increase_date = month_of_change[greatest_increase_pl]
greatest_decrease_date = month_of_change[greatest_decrease_pl]  

#print('Financial Analysis')
#print('----------------------------')
#print(f'Total Months: {line_num}')
#print(f'Total: ${total_net}')
#print(f'Average Change:: ${average_pnl}') 
#print(f'Greatest Increase in Profits:{greatest_increase_date} ${greatest_increase}')
#print(f'Greatest Decrease in Profits:{greatest_decrease_date} ${greatest_decrease}')



with open(budget_data_out, 'w') as file:
    
    file.write('Financial Analysis\n')
    file.write('----------------------------\n')
    file.write(f'Total Months: {line_num}\n')
    file.write(f'Total: ${total_net}\n')
    file.write(f'Average Change:: ${average_pnl}\n') 
    file.write(f'Greatest Increase in Profits:{greatest_increase_date} ${greatest_increase}\n')
    file.write(f'Greatest Decrease in Profits:{greatest_decrease_date} ${greatest_decrease}\n')

```
